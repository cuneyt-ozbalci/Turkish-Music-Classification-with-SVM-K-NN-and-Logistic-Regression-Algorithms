```python
 
%matplotlib inline
import numpy as np
import pandas as pd
import os
import pathlib
import csv
import librosa

#Veri setine ait parçaların bulunduğu dizindeki dosyaları listeliyoruz  
os.listdir('/content/drive/MyDrive/TezVeriSetiAsıl')

#Veri setinden örnek bir parça ataması yapıyoruz
sarkiAdi = f'/content/drive/MyDrive/TezVeriSetiAsıl/genres/klasik/klasik1.wav'
 
#örnek parçayı bir zaman serisi olarak yüklüyoruz, y zaman serisini sr ise örneklem oranını ifade etmekte. Uzunluğu 30 saniye olarak belirliyoruz 
y,sr = librosa.load(sarkiAdi,mono=True, duration=30)

#çıkarılacak öznitelikler için her bir sütun için ilgili özniteliğin adını headere atıyoruz
#(etiketler ve dosya isimleri daha sonra çıkarılacak yalnızca sayısal değerlere sahip işlenecek öznitelikler kalacak) 
header = 'filename chroma_sft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'

#mfcc1 den mfcc20'ye kadar 20 adet MFCCs değeri çıkaracağımız için 20'ye kadar mfcc etiketi oluşturuyoruz 
for i in range(1,21):
  header += f' mfcc{i}'
 
header += ' label'
header = header.split()
header
 
#csv dosyayı oluşturup bu aşamada özniteliklerin etiketlerini bu dosyaya aktarıyoruz ve türlerin etiketlerini oluşturuyoruz
file = open('ozniteliklerGuncel.csv', 'w', newline='')
 
with file: 
  writer = csv.writer(file)
  writer.writerow(header)
 
pd.DataFrame(pd.read_csv('ozniteliklerGuncel.csv')) 
 
genres = 'arabesk dini klasik pop rock rap'.split()
count=0

#Burada ise tüm parçalar için tek tek 7 özniteliği librosa kütüphanesi ile elde ediyoruz  
for genre in genres:
 
  for filename in os.listdir(f'genres/{genre}'):
    sarkiAdi = f'genres/{genre}/{filename}'
 
    y,sr = librosa.load(sarkiAdi, duration=30)
 
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    rmse = librosa.feature.rms(y=y)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw =   librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)
 
    to_append = f'{filename} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'
    for e in mfcc:
      to_append += f' {np.mean(e)}'
    to_append += f' {genre}'
 
    count +=1
    print(f'{count} - {to_append}')
 
    file = open('ozniteliklerGuncel.csv', 'a', newline='')
    
    with file:
      writer = csv.writer(file)
      writer.writerow(to_append.split())
```
